Problem Statement

The objective of this project is to build a machine learning model that predicts health insurance premiums based on an individual’s demographic, lifestyle, and medical attributes.

Accurate premium prediction helps insurance companies:

Price policies fairly

Assess medical risk

Automate premium calculation at scale

The dataset contains 1 million records sourced from a SQLite database and includes features such as age, BMI, smoking status, medical history, occupation, and insurance coverage level.

Business Objective

To predict the insurance charges (premium amount) for individuals using historical data and deploy the model in a form that can be used for batch or real-time predictions.

Target Variable

charges — Medical insurance premium amount (continuous)

Dataset Split Strategy

700,000 records → Training

200,000 records → Evaluation

Remaining records → Production / Live prediction
